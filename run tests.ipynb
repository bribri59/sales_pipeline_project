```python
ARCHIVE_ROOT = "/Volumes/workspace/sales/archive"
RAW_ROOT = "/Volumes/workspace/sales/raw_data"

boutiques = [
    "boutique_new_york",
    "boutique_paris",
    "boutique_tokyo",
]

print("=== RESTAURATION DES FICHIERS ARCHIVÉS ===")

for boutique in boutiques:
    archive_path = f"{ARCHIVE_ROOT}/{boutique}"
    raw_path = f"{RAW_ROOT}/{boutique}"

    print(f"\n📦 Boutique : {boutique}")
    print(f"Archive  -> {archive_path}")
    print(f"Raw data -> {raw_path}")

    archived_files = dbutils.fs.ls(archive_path)


    if not archived_files:
        print("Aucun fichier à restaurer.")
        continue

    dbutils.fs.mkdirs(raw_path)

    for f in archived_files:
        if not f.path.lower().endswith(".csv"):
            continue

        file_name = f.path.split("/")[-1]
        dest_path = f"{raw_path}/{file_name}"

        existing_files = [x.path.split("/")[-1] for x in dbutils.fs.ls(raw_path)]
        if file_name in existing_files:
            print(f"⏭️  {file_name} existe déjà dans raw → ignoré")
            continue

        print(f"↩️  Restauration : {file_name}")
        dbutils.fs.mv(f.path, dest_path)

print("\n✅ Restauration terminée")
```
